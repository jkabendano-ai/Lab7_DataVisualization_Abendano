```python
import pandas as pd
import matplotlib.pyplot as plt

# Student info
student_name = "Justine Kyle B. Abendaño"
student_id = "20250795"

# Correct file path
csv_file = r"C:\Users\jkabe\OneDrive\Documents\Lab7_DataVisualization_Abendano\spotify_top_1000_tracks.csv"

# Bar color
color_bar = "orange"

try:
    # Load dataset
    df = pd.read_csv(csv_file)

    # 🔥 Prevent freezing (safe sample)
    df = df.sample(n=min(500, len(df)), random_state=42)

    # Ensure duration_min exists
    if 'duration_min' not in df.columns and 'duration_ms' in df.columns:
        df['duration_min'] = df['duration_ms'] / 60000

    # Get top 10 longest tracks
    top_10 = df.sort_values(by='duration_min', ascending=False).head(10)

    # Display table
    print("\n--- TOP 10 LONGEST TRACKS ---")
    print(top_10[['track_name', 'artist', 'duration_min']])

    # Plot
    plt.figure(figsize=(10, 6))
    plt.barh(
        top_10['track_name'],
        top_10['duration_min'],
        color=color_bar,
        edgecolor='black'
    )

    plt.xlabel('Duration (Minutes)')
    plt.ylabel('Track Name')
    plt.title(f"Top 10 Longest Songs\n{student_name} ({student_id})")

    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print(f"Error: File not found → {csv_file}")
```
