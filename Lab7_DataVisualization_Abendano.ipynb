```python
import pandas as pd

# Load the dataset
dataset_path = r"C:\Users\jkabe\OneDrive\Documents\Lab7_DataVisualization_Abendano\spotify_top_1000_tracks.csv"
df = pd.read_csv(dataset_path, encoding="utf-8")

# Basic Preprocessing
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['year'] = df['release_date'].dt.year
```

```python
# Snippet 1

# Snippet 1
plt.figure(figsize=(10, 6))

# Generate the Histogram
df['duration_min'].plot(kind='hist', bins=30, color=color1, alpha=0.7, edgecolor='black')

# Add Labels and Title
plt.title(f"{student_name} ({student_id})\nHistogram of Song Duration")
plt.xlabel('Duration (min)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Snippet 2
import seaborn as sns
import matplotlib.pyplot as plt

# Make sure year exists (based on your earlier setup)
if 'release_year' in df.columns:
    df['year'] = df['release_year']
elif 'year' not in df.columns and 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    df['year'] = df['release_date'].dt.year

# Create decade safely
df['decade'] = (df['year'] // 10) * 10

# Drop missing values
plot_df = df.dropna(subset=['decade', 'popularity'])

# Boxplot
sns.boxplot(
    x='decade',
    y='popularity',
    data=plot_df,
    palette='coolwarm'
)

plt.title(f"{student_name} ({student_id})\nBoxplot of Popularity by Decade")
plt.xlabel("Decade")
plt.ylabel("Popularity")

plt.show()
```

```python
# Snippet 3
import seaborn as sns
import matplotlib.pyplot as plt

# Get top 10 artists safely
top_artists = df['artist'].value_counts().head(10).index

# Plot
sns.countplot(
    y='artist',
    data=df,
    order=top_artists,
    palette='viridis'
)

plt.title(f"{student_name} ({student_id})\nTop 10 Artists by Song Count")
plt.xlabel("Number of Songs")
plt.ylabel("Artist")

plt.show()
```

```python
# Snippet 4
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure duration_min exists
if 'duration_ms' in df.columns:
    df['duration_min'] = df['duration_ms'] / 60000

# Ensure decade exists
if 'decade' not in df.columns:
    if 'year' not in df.columns and 'release_date' in df.columns:
        df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
        df['year'] = df['release_date'].dt.year

    df['decade'] = (df['year'] // 10) * 10

# Clean data
plot_df = df[['duration_min', 'decade']].dropna()

# SAFE sampling (fix crash)
sample_size = min(1000, len(plot_df))
plot_df = plot_df.sample(n=sample_size, random_state=42)

# Plot
plt.figure(figsize=(10, 6))

sns.violinplot(
    data=plot_df,
    x='duration_min',
    y='decade',
    hue='decade',
    palette='coolwarm',
    density_norm='width',
    legend=False
)

plt.title(f"{student_name} ({student_id})\nViolin Plot of Duration by Decade")
plt.xlabel("Duration (min)")
plt.ylabel("Decade")

plt.show()
```

```python
# Snippet 5
import matplotlib.pyplot as plt

# Make sure decade exists
if 'decade' not in df.columns:
    if 'year' not in df.columns and 'release_date' in df.columns:
        df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
        df['year'] = df['release_date'].dt.year

    df['decade'] = (df['year'] // 10) * 10

# Clean data
plot_df = df.dropna(subset=['decade', 'popularity'])

# Average popularity per decade
avg_pop = plot_df.groupby('decade')['popularity'].mean()

# Plot
avg_pop.plot(
    kind='line',
    color=color1,
    marker='o'
)

plt.title(f"{student_name} ({student_id})\nAverage Popularity per Decade")
plt.xlabel("Decade")
plt.ylabel("Average Popularity")

plt.grid(True)
plt.show()
```

```python
# Snippet 6
import matplotlib.pyplot as plt

# Create release year safely
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year

# Count songs per year
count_by_year = df['release_year'].value_counts().sort_index()

# Plot
count_by_year.plot(
    kind='area',
    color=color2,
    alpha=0.7
)

plt.title(f"{student_name} ({student_id})\nNumber of Songs Over Time")
plt.xlabel("Year")
plt.ylabel("Count")

plt.grid(True)
plt.show()
```

```python
# Snippet 7
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure duration_min exists
if 'duration_ms' in df.columns:
    df['duration_min'] = df['duration_ms'] / 60000

# Ensure decade exists
if 'decade' not in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    df['year'] = df['release_date'].dt.year
    df['decade'] = (df['year'] // 10) * 10

# Clean data
plot_df = df.dropna(subset=['decade', 'duration_min', 'popularity'])

# Facet Grid
g = sns.FacetGrid(plot_df, col='decade', col_wrap=4, height=3)

g.map_dataframe(
    sns.scatterplot,
    x='duration_min',
    y='popularity',
    color=color1
)

g.fig.suptitle(
    f"{student_name} ({student_id})\nPopularity vs Duration per Decade",
    y=1.02
)

plt.show()
```

```python
# Snippet 8
import matplotlib.pyplot as plt

# Get top 10 longest songs
longest = df.nlargest(10, 'duration_min')

plt.figure(figsize=(10, 6))

plt.stem(
    longest['track_name'],
    longest['duration_min'],
    linefmt='#888888',
    markerfmt='o',
    basefmt=" "
)

plt.xticks(rotation=90)

plt.title(f"{student_name} ({student_id})\nTop 10 Longest Songs")
plt.ylabel("Duration (min)")

plt.tight_layout()
plt.show()
```

```python
# Snippet 9
import matplotlib.pyplot as plt

# Compute average duration per artist
avg_duration = df.groupby('artist')['duration_min'].mean().nlargest(5)

plt.figure(figsize=(8, 5))

# Correct plotting (y = artist, x = duration)
plt.plot(
    avg_duration.values,
    avg_duration.index,
    'o',
    color=color2
)

plt.title(f"{student_name} ({student_id})\nAverage Duration: Top 5 Artists")
plt.xlabel("Duration (min)")
plt.ylabel("Artist")

plt.grid(True)

plt.show()
```

```python
# Snippet 10
import pandas as pd
import matplotlib.pyplot as plt

# Create crosstab
crosstab = pd.crosstab(df['decade'], df['artist'])

# Get top 3 artists
top3 = df['artist'].value_counts().head(3).index

# Keep only top 3 safely (avoid KeyError)
crosstab_top3 = crosstab.loc[:, crosstab.columns.intersection(top3)]

# Plot
crosstab_top3.plot(
    kind='bar',
    stacked=True,
    colormap='coolwarm',
    figsize=(10, 6)
)

plt.title(f"{student_name} ({student_id})\nTop 3 Artists by Decade")
plt.xlabel("Decade")
plt.ylabel("Number of Songs")

plt.show()
```
